```python
import pandas as pd
import numpy as np

# Load the CSV
df = pd.read_csv('C:/[your directory here]/EPF_cluster_1_DistMat.csv', index_col=0)

# Convert to NumPy array
distance_matrix = df.values
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.spatial.distance import squareform

# Force symmetry by averaging with its transpose
distance_matrix = (distance_matrix + distance_matrix.T) / 2

# Convert square distance matrix to condensed form
condensed_dist = squareform(distance_matrix)

# Compute linkage
Z = linkage(condensed_dist, method='average')  # You can try 'ward', 'complete', etc.
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram

# Plot
plt.figure(figsize=(18, 8))
dendrogram(Z, labels=df.index.tolist(), leaf_rotation=90)
plt.title('Hierarchical Clustering Dendrogram of EPF Proteins')
plt.xlabel('Protein')
plt.ylabel('Distance')
plt.tight_layout()
plt.show()

plt.savefig('C:/[your directory here]/EPF_dendrogram.png', dpi=300, bbox_inches='tight')
```
