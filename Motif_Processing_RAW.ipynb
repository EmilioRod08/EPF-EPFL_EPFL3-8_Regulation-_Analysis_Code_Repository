```python
import pandas as pd

#Load CSV file
#First column is gene names, remaining columns are receptors
df = pd.read_csv("C:/Users/Emili/Downloads/EPFL3_PLACEmotif_counts.csv")

# Set gene names as index (optional, for clarity)
df = df.set_index(df.columns[0])

# Convert all positive integers to 1, keep 0 as 0
binary_df = (df > 0).astype(int)

# Calculate coverage per receptor (mean of 0/1 values)
# Since values are now binary, the column mean == proportion of genes with ≥1
coverage = binary_df.mean(axis=0)

# Sort receptors by highest coverage
coverage_sorted = coverage.sort_values(ascending=False)

print("Receptor coverage proportion (1 = all genes have ≥1):")
print(coverage_sorted)

# If you just need the top receptor(s)
top_receptors = coverage_sorted[coverage_sorted == coverage_sorted.max()]
print("\nTop receptor(s):")
print(top_receptors)
```
